## 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.


```python
def sumAll(l):
    sum = 0 
    for elem in l : 
        sum += elem
    return sum
```

## 4. Write a Python function that checks if a given number is prime or not from 1 to 200.



```python
def prime(x):
    if x <= 1 :
        return False
    for i in range(2 , x-1):
      
        if x % i == 0 :
            return False
    return True
    
```

```python
primes =[]
for i in range (1, 201):
    if prime(i):
        primes.append(i)
```

## 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.


```python
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms  # Number of terms in the sequence
        self.current_term = 0   # Current term position
        self.a, self.b = 0, 1   # Initial two numbers of the Fibonacci sequence

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_term >= self.n_terms:  # Stop iteration when terms exceed specified limit
            raise StopIteration
        
        if self.current_term == 0:
            self.current_term += 1
            return self.a
        elif self.current_term == 1:
            self.current_term += 1
            return self.b
        
        # Calculate the next Fibonacci number
        next_fib = self.a + self.b
        self.a, self.b = self.b, next_fib
        self.current_term += 1
        return next_fib

# Example usage
fib = FibonacciIterator(10)
for num in fib:
    print(num)
```

## 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.


```python
def pow_2 (x):
    for i in range(1,x +1):
        yield 2 ** i
```

## 7. Implement a generator function that reads a file line by line and yields each line as a string.


```python
import os 
def reader(filename):
    
    file_path = "./" + filename
    if  not os.path.exists(file_path):
        return "The file do not exist"
    with open(file_path) as file_obj: # default is the 'r' mode of the file 
        for line in file_obj:
            # this is same as file_obj.readlines() which return a list of all the lines
            
            yield line.strip() # this removes the extra newline character 
    
```

```python

x = reader("textile.txt")
```

```python
str(x)
```

```python
for line in x:
    print(line)
    
```

## 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.


`°F = (°C × 9/5) + 32` formula for converting celcius to Farehheit
`°C = (°F − 32) x 5/9 ` formula for converting Farenheit to celcius.

```python
def tempFahren(temps):
    ''' converts the celcius temperature to degree Fahrenheit and returns a list of converted temperatures to Fahreheit'''
    # round function rounds the deicmal value up to 2 decimals ceil or floor it decides accordingly
    return list(map(lambda x  : round((x * 1.8) +  32,2), temps))
    
    
```

```python
celsius_temps = [0, 20, 37, 100, -40]
```

```python
tempFahren(celsius_temps)
```
